```python
import pandas as pd
```

```python
score = pd.read_csv('score.csv', encoding = 'euc-kr' , index_col = '과목')
score
```

```python
import numpy as np
np.sum(score)
```

```python
# 학급결 과목 점수 총계
# 순위에 맞게 출력
score.sum()
score.sum().sort_values(ascending = False)
```

```python
# 과목별 점수 총계
np.sum(score, axis=1)
score.sum(axis=1)
```

```python
# score['합계'] = [201,282,253,298,163]
# np.sum(score.loc[:,:'4반'], axis = 1)
score['합계'] = score.loc[:,:'4반'].sum(axis=1)
score
```

```python
score['평균'] = score['합계'] / 4
# sum 대신 mean으로 변경
# np.mean(score.loc[:,:'4반'], axis = 1)
# score['평균'] = score.loc[:,:'4반'].mean(axis=1)
score
```

```python
# 인덱스를 추가하는 방법
score.loc['반평균'] = score.loc[:'과학',:].mean()
score
```

```python
# 과목별로 가장 큰값과 가장 작은값의 차이를 구하자
# 가장 큰값 구하기
# max()
max_score = score.loc[:'과학',:'4반'].max(axis=1)
print(max_score)
# 가장 작은 값 구하기
# min()
min_score = score.loc[:'과학',:'4반'].min(axis=1)
min_score
```

```python
max_score - min_score
```

## APPLY()

```python
# apply()
# 행이나 열 단위로 복잡한 처리를 하고 싶을 때 사용
# 적용시킬 데이터에 행 / 열 단위로 함수를 적용시켜줌
# (적용 시킬 데이터).apply(처리할 작업, 기준)
# 행기준 axis = 1
```

```python
def max_min(x):
    return x.max() - x.min()
```

```python
score.loc[:'과학',:'4반'].apply(max_min , axis = 1)
```

```python
# 결측치를 채우는 함수
# fillna()
data_dic = {'A':[1,2,3,4,4], 'B':[1,2,2,3,3], 'C':[1,2,4,4,5]}
df = pd.DataFrame(data_dic)
df
```

```python
# 해석 : A에 1이 1개, 2가 1개, 3이 1개, 4개 2개, 5개 0개 다
df1 = df.apply(pd.value_counts)
df1
```

```python
# 결측치를 채우는 함수
# fillna()
# 괄호 안의 수로 NaN을 채움
# 값을 출력하고 저장하지 않음
df1 = df1.fillna(100)
df1
```

## 카테고리 데이터 만들기

```python
# 카테고리 데이터 만들기
# cut(데이터, 기준, 라벨)
```

```python
ages = [0,2,10,21,23,37,31,61,20,41,32,100]
bins = [0,15,25,35,60,99] # 구분할 기준 0 ~ 15 : 0초과 15이하
labels = ['미성년자','청년','중년','장년','노년'] # 구분별 들어갈 이름
```

```python
pd.cut(x = ages, bins = bins , labels = labels)
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4, 5, 6, 7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                   index=[8, 9, 10, 11])
```

```python
# 데이터 프레임 합치기
# concat()
# 저장되지 않는다, 적은 순서대로 합쳐짐
result = pd.concat([df1,df2,df3])
result
```

```python
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'], 
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                   index=[2, 3, 6, 7])
```

```python
# axis = 0 : 컬럼을 기준으로 
pd.concat([df1,df4], axis = 1) # 인덱스를 
# 인덱스가 없을경우 앞에서부터 순서대로 정리 후
# 없는 데이터는 NaN
```

```python
# 같은 인덱스를 가지고 있는 값만 출력
pd.concat([df1,df4], axis = 1, join = 'inner')
```

```python
# 인덱스(컬럼) 초기화
pd.concat([df1,df4], axis = 1, ignore_index = True )
```

```python
# 인덱스 초기화
pd.concat([df1,df4], ignore_index = True )
```

```python
df5 = pd.DataFrame({'key': ['K0','K2','K3','K4'],
'A':['A0','A1','A2','A3'],
'B':['B0','B1','B2','B3']})


df6 = pd.DataFrame({'key': ['K0','K1','K2','K3'],
'C':['C0','C1','C2','C3'],
'D':['D0','D1','D2','D3']})

print(df5)
df6
```

```python
# merge
# 데이터를 기준으로 df 합침
# 기준이 같은것만 출력 : how = inner
# 한번에 최대 2개만 합칠수있음
pd.merge(df5,df6,on = 'key')
# 기준 컬럼 데이터를 기준으로 합쳐줌
# 공통적으로 가지고 있는 데이터를 가져와 합쳐준다
```

```python
# 모든 데이터 출력
pd.merge(df4,df6, how='outer')
```

```python
# how : left, right
# 적은 대상의 왼쪽것에 기준
pd.merge(df5,df6,how='left')
```

```python
# how : left, right
pd.merge(df5,df6,how='right')
```
